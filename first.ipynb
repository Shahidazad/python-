```python
name=input('Enter file:')
handle=open(name,'r')
```

```python
# count word frequency
counts=dict()
for line in handle:
    words=line.split()
    for word in words:
        counts[word]=counts.get(word,0) +1
```

```python
#  find the most common words
bigcount=None
bigword=None
for word,count in counts.items():
    if bigcount is None or count > bigcount:
        bigcount=count
        bigword=word
print(bigword,bigcount)
```
